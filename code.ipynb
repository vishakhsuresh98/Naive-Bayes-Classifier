```python
# Binning (to handle real valued features for computing conditional probs.)

no_of_bins = 30
bins = np.zeros((no_of_bins, d))# 2 * d)) # row1 - min and row2 - max

for i in range(0, d):
    col_label = 'f' + str(i+1)
    column = data[col_label]
    
    # finding max and min in each column for binning
    min = max = column[0]
    for j in range(1, len(column)):
        if column[j] < min:
            min = column[j]
        if column[j] > max:
            max = column[j]        
    
    # splitting range into 20 bins
    d = (max - min) / no_of_bins
    for j in range(0, no_of_bins):
        bins[j, i] = min + j * d
#         bins[j, 2*i] = min + j * d
#         bins[j, 2*i + 1] = min + (j+1) * d
```

```python
bins
```

```python
def find_bin(val, feature):
    bin = 0
    for i in range(0, no_of_bins):
        if val > bins[i, feature]:
            bin = i
    return bin       
```

```python
# relabelling the feature values (Binnning)

for i in range(0, data.shape[1]):
    col_label = 'f' + str(i+1)
    for j in range(0, data.shape[0]):
        data.loc[j, col_label] = int(find_bin(data.loc[j, col_label], i))
```

```python
import math, random

def train_test_split(X, y, f):
    # f : fraction of samples taken for training; in the range (0,1)
    n = X.shape[0]
    m = math.floor(f * n)
    row_indices = [i for i in range(0, X.shape[0])]

    random.shuffle(row_indices)
    test_rows = row_indices[0:n-m]

    train_rows = [i for i in range(0, m)]
    
    X_test = pd.DataFrame(columns = labels[0:len(labels)-1])
    y_test = pd.DataFrame(columns = ['t'])
    
    for row in test_rows:
        X_test.loc[len(X_test)] = X.iloc[row]
        y_test.loc[len(y_test)] = y.iloc[row]
    
    X_train = pd.DataFrame(X.drop(X.index[test_rows]))
    y_train = pd.DataFrame(y.drop(y.index[test_rows]))
    
    return X_train, X_test, y_train, y_test
```

```python
def compute_prob_y():
    prob_1 = 0
    prob_0 = 0
    for i in range(0, X_train.shape[0]):
        if target.loc[i, 't'] == 1:
            prob_1 += 1
        else:
            prob_0 += 1
            
    prob_1 /= (X_train.shape[0] + 0.0)
    prob_0 /= (X_train.shape[0] + 0.0)
#     print("prob one : ", p_1)
#     print("prob zero : ", p_0)
    return prob_1, prob_0
   
prob_1, prob_0 = compute_prob_y()
```

```python
def compute_prob_x_given_1():
    prob_x_given_1 = np.zeros((no_of_bins, X_train.shape[1]))

    for i in range(0, X_train.shape[0]):
        for j in range(0, X_train.shape[1]):
            x_val = int(data.loc[i, 'f' + str(j+1)])
            y_val = target.loc[i, 't']
            if y_val == 1:
                prob_x_given_1[x_val, j] += (1.0 / (prob_1 * X_train.shape[0]))
                
    return prob_x_given_1

prob_x_given_1 = compute_prob_x_given_1()
```

```python
prob_x_given_1
```

```python
def compute_prob_x_given_0():
    prob_x_given_0 = np.zeros((no_of_bins, X_train.shape[1]))

    for i in range(0, X_train.shape[0]):
        for j in range(0, X_train.shape[1]):
            x_val = int(data.loc[i, 'f' + str(j+1)])
            y_val = target.loc[i, 't']
            if y_val == 0:
                prob_x_given_0[x_val, j] += (1.0 / (prob_0 * X_train.shape[0]))
                
    return prob_x_given_0

prob_x_given_0 = compute_prob_x_given_0() 
```

```python
prob_x_given_0
```

```python
def predict(x):
    prob_1_given_x = prob_1
    for i in range(0, X_train.shape[1]):
        x_i = int(x[i])
        prob_1_given_x *= prob_x_given_1[x_i, i]
    prob_0_given_x = prob_0
    for i in range(0, X_train.shape[1]):
        x_i = int(x[i])
        prob_0_given_x *= prob_x_given_0[x_i, i]
    
    if prob_0_given_x < prob_1_given_x:
        pred = 1
    else:
        pred = 0
    return pred      
```

```python
# Accuracy = # correct predictions / # predictions

correct_preds = 0

for i in range(0, X_test.shape[0]):
    x = X_test.iloc[i]
    if predict(x) == y_test.loc[i, 't']:
        correct_preds += 1

accuracy = correct_preds / (X_test.shape[0] + 0.0)        
```
